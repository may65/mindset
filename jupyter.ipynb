```python
#количество ребер(больше 1) на участника 1
for index1,count1 in dt['ФИО участника события 1'].value_counts().items():
    if count1 > 1:
        print(index1,count1)
        for i, r in dt.iterrows():
            if index1 == r['ФИО участника события 2']:
                print("\t",r['id события'],r['ФИО участника события 1'],r['ФИО участника события 2'])
```

```python
#количество ребер(больше 1) на участника 2
for index2,count2 in dt['ФИО участника события 2'].value_counts().items():
    if count2 > 1:
        print(index2,count2)
        for i, r in dt.iterrows():
            if index2 == r['ФИО участника события 1']:
                print("\t",r['id события'],r['ФИО участника события 1'],r['ФИО участника события 2'])
```
